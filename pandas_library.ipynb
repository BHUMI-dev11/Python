Inspecting a DataFrame

```python
#getting last 5 rows by default
diabetes_df.tail()
```

```python
#information about DataFrame
diabetes_df.info()
```

```python
#finding the number of missing values
diabetes_df.isnull().sum()
```

```python
#counting the values based on the labels
diabetes_df.value_counts('Outcome')
```

```python
diabetes_df.value_counts('Age')
```

Statistical Measures

```python
#count or number of values
diabetes_df.count()
```

```python
#mean value - column wise
diabetes_df.mean()
```

```python
#standar deviation - column wise
diabetes_df.std()
```

```python
#minimum value
diabetes_df.min()
```

```python
#maximum value
diabetes_df.max()
```

Manipulating a DataFrame

```python
#adding a column to a dataframe
diabetes_df['Current_Condition'] = diabetes_df.Outcome
```

```python
diabetes_df.head()
```

```python
#removing a particular row
diabetes_df.drop(index=1, axis=0)
```

```python
#drop a column
diabetes_df=diabetes_df.drop(columns='Current_Condition', axis=1)
```

```python
#locating a row using the index value
diabetes_df.iloc[2]
```

```python
#locating a particular column
print(diabetes_df.iloc[:,0]) #first column
```
