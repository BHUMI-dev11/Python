Seaborn : Data Visulatization Library

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
#total bills vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
# visulaize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size = 'size')
```

```python
#load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Scatter Plot
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Count Plot
sns.countplot(data=titanic,x='sex')
```

```python
sns.countplot(data=titanic,x='survived')
```

```python
#Bar Chart
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```

```python
#distribution plot
sns.displot(data=titanic,x='age')
```

```python
#constructing a heatmap
flight1 = sns.load_dataset('flights')
flight1.head()
```

```python
flight1 = flight1.drop('month',axis=1)
flight1.head()
```

```python
#heatmap
sns.heatmap(flight1.corr(),annot=True)
```
