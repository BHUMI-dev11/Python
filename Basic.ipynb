```python
def quad_check(points):
  match points :
    case (1,2):
      return "Points are in first quadrant"
    case (-1,2):
      return "Points are in second quadrant"
    case _:
      return "default case"
print(quad_check((3,5)))
quad_check((-1,2))
```

#OOPs in Python
OOPs is a programming paradigm that organizes code into objects that contain both data(attributes) and behavior(methods).

*Key Concepts of OOPs:*

Class,Object,Attribute,Methods,Encapsulation,Inheritance,Polymorphism

__init__() is called the constructor.

1.  It executes automatically whenever
an object is created.

2. Its job is to initialize the object's data.

**Parameters** : self

Refers to the current object.
It is automatically passed by Python.

```python
# Creating a Class
class Car:
  def __init__(self,brand,model):
    self.brand = brand
    self.model = model
  def display_info(self):
    return f"{self.brand},{self.model}"

#Creating an Object(Instance)
car1 = Car("Toyoto", "Camry")
print(car1.display_info())
```

```python
#Encapsulation(Data Hiding) -> using getter and setter
class BankAccount:
  def __init__(self, balance):
          self.balance = balance  # Private Attribute
  def get_balance(self): # Getter
    return self.balance
  def deposit(self, amount): # Setter
    if amount > 0:
              self.balance += amount
# Using Encapsulation
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance()) # Output: 1500
```

```python
#Single Inheritance
class Animal:
  def speak(self):
    return "Animal makes a sound"
class Dog(Animal):
  def speak(self):
    return "Bark"
dog = Dog()
print(dog.speak())
```

```python
#Multiple Inheritance
class A:
  def method_a(self):
    return "Method A"
class B:
  def method_b(self):
    return "Method B"
class C(A,B):
  pass
obj = C()
print(obj.method_a())
print(obj.method_b())
```

```python
#Polymorphism
class Bird:
  def fly(self):
    return "Birds can fly"
class Penguin(Bird):
  def fly(self):
    return "Penguins cannot fly"
bird = Bird()
penguin = Penguin()
print(bird.fly())
print(penguin.fly())
```

```python
#Abstraction
from abc import ABC, abstractmethod
class Shape(ABC):
  def area(self):
    pass
class Square(Shape):
  def __init__(self,side):
    self.side = side
  def area(self):
    return self.side * self.side
square = Square(4)
print(square.area())
```

#List Comprehension
It allows you to generate lists in a single line of code, making code more readable and Pythonic.

[expression for item in iterable]

```python
evens = [x for x in range(10) if x%2==0]
print(evens)
```

```python
#Using if-else Condition in List Comprehension
numbers = ["even" if x%2==0 else "odd" for x in range(5)]
numbers
```

```python
#Nested Loops in List Comprehension
pairs=[(x,y) for x in range(2) for y in range(3)]
pairs
```

```python
#List Comprehension with Functions
words=["hello" ,"world" , "pyhton"]
upper_words = [word.upper() for word in words]
upper_words
```

```python
#List Comprehension with Nested List Comprehension
matrix = [[1,2],[3,4],[5,6]]
flattened = [num for row in matrix for num in row]
print(flattened)
```

```python
#List Comprehension with Set and Dictionary Comprehensions
unique_num = {x for x in[1,2,3,2,4,4]}
print(unique_num)

squared_dict = {x : x**2 for x in range(5)}
print(squared_dict)
```
