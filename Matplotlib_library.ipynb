Matplotlib:
1. Useful for making plots


```python
#importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for our plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
#Plotting the data
#sine wave
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Triogometric Functions')

plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Parabola')

plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #as we are creating figure therefore we need to mention the axes
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('Languages')
plt.ylabel('Number of students')
plt.title('Bar Plot')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax1.pie(students,labels=languages,autopct='%1.2f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y)
plt.show()
```

3D Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 2* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.plot3D(x,y,z,'green')
plt.show()
```
