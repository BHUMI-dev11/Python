```python
import numpy as np
```

List v/s Numpy -> Time Taken

```python
from time import process_time
```

```python
python_list = [i for i in range(10000)]
start_time = process_time()
python_list = [i+5 for i in python_list]
end_time = process_time()
print(end_time - start_time)
```

```python
num_array = np.array([i for i in range(10000)])
start_time = process_time()
num_array += 5
end_time = process_time()
print(end_time - start_time)
```

Numpy Array

```python
#creating 1D array
np_array = np.array([1,2,3,4,5])
print(np_array)
print(np_array.shape)
type(np_array)
```

```python
#creating multidimensional array
b = np.array([(1,2,3),(5,6,7)])
print(b)
print(b.shape)
```

```python
c = np.array([1,2,4,5,6,7,3],dtype = float)
c
```

Initial Placeholders in numpy array

```python
#create a numpy array of zeros
x =np.zeros((4,5))
print(x)
```

```python
#create a numpy array of ones
y = np.ones((3,3))
print(y)
```

```python
#array of a particular value
z = np.full((5,4),5)
print(z)
```

```python
#create an identity matrix
a = np.eye(4)
print(a)
```

```python
#create a numpy array with random values --->by default it gives the value between zero and one
b = np.random.random((3,4))
print(b)
```

```python
#random integer values array in a specific range
c = np.random.randint(10,100,(3,5))
print(c)
```

```python
#create an array of evenly spaced values -->specifying the number of values requires
d= np.linspace(10,30,6)
print(d)
```

```python
#create an array of evenly spaced values -->specifying the step
e = np.arange(10,30,5)
print(e)
```

```python
#convert a list to a numpy array
list2 = [10,20,20,20,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

Mathematical Operation on Numpy Array

```python
list1 = [1,2,3,4,5]
list2 = [5,6,7,8,9]
print(list1 + list2) #concatenate or joins two list
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)

b= a.reshape(3,2)
print(b)
print(b.shape)
```
